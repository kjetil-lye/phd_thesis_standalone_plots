```python
import numpy as np
import sys
sys.path.append('../../python')
import plot_info
import matplotlib.pyplot as plt
plot_info.set_notebook_name('Perturbation.ipynb')
```

```python

K = 10
x_K = np.linspace(0, 1, K)
a = np.sin(10*np.pi*x_K+0.13)
b = np.cos(10*np.pi*x_K+0.9)


x = np.linspace(0, 1, 1024)

y = np.zeros_like(x)

for k in range(K):
    y += (a[k]/np.linalg.norm(a, ord=1))*np.cos(2*np.pi*k*(x+2*np.pi*b[k]))
    
plt.plot(x, y)
plt.ylim([-1, 1])
plt.xlabel('$x$')
plt.ylabel(f'$\\sum_{{k=1}}^{{{K}}}\\frac{{a_k}}{{||a||_{{\\ell^1}}}}\\sin(2\pi k(x+b_k))$')
plt.title(f"$I_{{{K}}}(a,b,x)$ for $a_k=\\sin(10\\pi (k-1)/{{{K-1}}}+0.13), b_k=\\sin(10\\pi (k-1)/{{{K-1}}}+0.9)$")
plot_info.showAndSave(f'example_standard_perturbation_{K}')
```
